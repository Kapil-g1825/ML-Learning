Assignment 3:- Shopping Cart System 

```python
cart = []
print("\n1. Add Product")
print("2. Remove Product")
print("3. View Products")
print("4. Exit")
while True:

    choice = input("Enter choice: ")

    if choice == "1":
        item = input("Enter product name: ")
        cart.append(item)

    elif choice == "2":
        item = input("Enter product to remove: ")
        if item in cart:
            cart.remove(item)
        else:
            print("Product not found.")

    elif choice == "3":
        print("Products:", cart)
        print("Total products:", len(cart))

    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid choice.")
```
