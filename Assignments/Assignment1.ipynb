Assignment 1:-Smart ATM Simulator.


Develop a Python program that simulates the basic operations of an ATM.

```python
balance = 10000
history = []
print("\n1. Check Balance")
print("2. Withdraw")
print("3. Deposit")
print("4. Exit")

while True:
    choice = input("Enter choice: ")
    if choice == "1":
        print("Balance:", balance)
    elif choice == "2":
        amount = int(input("Enter amount: "))
        if amount <= balance:
            balance -= amount
            history.append(f"Withdraw ₹{amount}")
            print("Money withdrawn.")
        else:
            print("Insufficient balance.")
    elif choice == "3":
        amount = int(input("Enter amount: "))
        balance += amount
        history.append(f"Deposit ₹{amount}")
        print("Money deposited.")
    elif choice == "4":
        print("Transaction History:", history)
        print("Thank you!")
        break

    else:
        print("Invalid choice.")
```
