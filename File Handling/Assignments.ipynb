Assignment 4:count words,characters and lines

```python
# write a program to count a frequency of word in a sentence
sentence=input("Enter a sentence")
words=sentence.split()
frequency={}

for word in words:
  if word in frequency:
    frequency[word]+=1
  else:
    frequency[word]=1
print("\n Word frequencies are as follows:")
for word,count in frequency.items():
   print(word,":",count)
```

Assignement 6. Count vowels in a file

```python
file = open("myfile.txt", "r")
text = file.read()
file.close()

count = 0
for ch in text:
    if ch.lower() in "aeiou":
        count += 1

print("Number of vowels:", count)
```

Assignment 10. Store student details

```python
file = open("students.txt", "a")
n = int(input("Enter number of students: "))
for i in range(n):
    roll = input("Roll No: ")
    name = input("Name: ")
    marks = input("Marks: ")
    file.write(roll + ", " + name + ", " + marks + "\n")
print("\nStudent Records:")
file = open("students.txt", "r")
print(file.read())
file.close()
```
