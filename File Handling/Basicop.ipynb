1.Write mode

```python
file=open("firstfile.txt","w")
file.write("hello every one")

file.close()
```

2.Read mode

```python
file=open("firstfile.txt","r")
print(file.read())
```

3.Append mode

```python
file=open("firstfile.txt","a")
file.write("This is append program")
file.close()
```
